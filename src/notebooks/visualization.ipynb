# Exam - Rest Client

```
import requests
import pandas as pd
import numpy as np
from itertools import cycle
host = 'https://05dea53b-e624-4a45-a48c-4f8a58ba823b.deepnoteproject.com'
list_endpoint = f'{host}/devices'
device_endpoint = f'{host}/device'
spy_endpoint = f'{host}/spy'
```

### We can retrive all the device that are actually monitoring someone

```
response = requests.get(list_endpoint)
if response.status_code == 200:
    body = response.json()
    print(body)
```

### Here we can make some statistic and see how the single device performs.

```
from datetime import datetime
```

```
response = requests.get(list_endpoint)
list_mac = []
if response.status_code == 200:
    body = response.json()
    list_mac = body['mac_addresses']
print(list_mac)
#df = pd.DataFrame(values, columns=['Datetime', 'Latency'])
#df
complete_list = []
for mac in list_mac:
    from_datetime = int(datetime.timestamp(datetime(2023, 6, 8)))*1000
    to_datetime = int(datetime.timestamp(datetime(2023, 6, 13)))*1000
    response = requests.get(device_endpoint + f'/{mac}?from={from_datetime}&to={to_datetime}')
    if response.status_code == 200:
        body = response.json()
        print(body)
        timestamps = body['timestamps']
        latencies = body['latencies']
        probabilities = body['probabilities']
        predictions = body['predictions']
        predictions = [int(x + 1) for x in predictions]

        zip_list = list(zip(cycle([mac]), timestamps, latencies, probabilities, predictions))
        complete_list.extend(zip_list)

df = pd.DataFrame(complete_list, columns=['Mac_Address','Datetime', 'Latency', 'Probability', 'Prediction'])
        
df
```
